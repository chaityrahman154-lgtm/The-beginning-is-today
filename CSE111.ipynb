```python
class MetroRail:
    def __init__(self,name,number):
        self.name=name
        self.number=number
        print(f"{self.name} has total {self.number} stations")
        self.dic={}
        self.station_count=1
    def board_passenger(self,*args):
        for i in range(len(args)):
            if i%2!=0:
                if args[i]<self.number:
                    self.dic[args[i-1]]=args[i]
                    print(f"{args[i-1]}, welcome to {self.name}")
                else:
                    print(f"Sorry {args[i-1]} the metro does not go to {args[i]}")
    def reach_next_station(self):
        self.x=False
        self.station_count+=1
        print(f"The train has arrived at station {self.station_count}")
        lst1=list(self.dic.keys())
        lst2=list(self.dic.values())
        self.s=""
        for i in range(len(lst1)):
            if lst2[i]==self.station_count:
                lst2.remove(lst2[i])
                lst1.remove(lst1[i])
                self.s+=lst1[i]+","
                self.x=True
        if self.x==True:
            print("Following passenger please get down")
            print(self.s[:(len(self.s)-1)])
        else:
            print("No passsenger to get down")
        lst=[]
        for i in range(len(lst1)-1):
            if lst2[i]==lst2[i+1]:
                lst.append(lst1[i])
                lst.append(lst1[i+1])
            else:
                print("Current passenger list")
                print(f"Station {lst2[i]}: {lst}")
                lst=[]

metro1=MetroRail("MRT Line 1",4)
print("1==================")
metro1.board_passenger("Alice",3,"Bob",3)
print("2==================")
metro1.reach_next_station()
print("3==================")
metro1.board_passenger("Harry",4)
print("4==================")
metro1.board_passenger("John",4)
print("5==================")
metro1.reach_next_station()
print("6==================")
metro1.board_passenger("Charlie",6)
print("7==================")
metro1.reach_next_station()

```

```python
#Task 3
class Department:
    def __init__(self,d="ChE",s=5):
        self.department=d
        print(f"The {d} Department has {s} sections.")
        self.section=s
        self.avg=0
        self.l=[]
    def add_students(self,*args):
        if self.section==len(args):
            self.avg=0
            for i in args:
                self.avg+=i
            self.avg/=len(args)
            print(f"The {self.department} Department has an average of {round(self.avg,2)} students in each section.")
        else:
            print(f"The {self.department} Department doesn't have {len(args)} sections.")
    def merge_Department(self,*args):
        total_students=self.avg*self.section
        for j in args:
            self.l.append(j)
            print(f"{j.department} Department is merged to {self.department}.")
        for i in self.l:
                total_students+=(i.avg*i.section)
        total_students/=self.section
        return f"Now the Engineering Department has an average of {total_students} students in each section."
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shape:
  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base

  def get_height_base(self):
    return "Height: "+str(self.height)+", Base: "+str(self.base)

class triangle(Shape):
    def calcArea(self):
        self.area=(1/2)*self.base*self.height
    def printDetail(self):
        return f"Shape name: {self.name}\n{self.get_height_base()}\nArea: {self.area}"


class trapezoid(Shape):
    def __init__(self,n,h,b,s):
        super().__init__(n,h,b)
        self.side=s
    def calcArea(self):
        self.area=(1/2)*(self.base+self.side)*self.height
    def printDetail(self):
        return f"Shape name: {self.name}\n{self.get_height_base()}, Side_A: {self.side}\nArea: {self.area}"

tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
class TwoDVector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add2DVectors(self, *vectors):
        for i in vectors:
            self.x+=i.x
            self.y+=i.y
    def print2DVector(self):
        if self.y >= 0:
            y = "+ "+str(self.y)
        else:
            y = str(self.y)
        print(f"{self.x}i {y}j")

class ThreeDVector(TwoDVector):
    def __init__(self,x,y,z):
        super().__init__(x,y)
        self.z=z
    def add3DVectors(self,*args):
        super().add2DVectors(*args)
        for i in args:
            self.z+=i.z
    def print3DVector(self):
        if self.y >= 0:
            y = "+ "+str(self.y)
        else:
            y = str(self.y)
        if self.z >= 0:
            z= "+ "+str(self.z)
        else:
            z = str(self.z)
        print(f"{self.x}i {y}j {z}k")
    def multiplyScalar(self,nmber):
        self.x*=nmber
        self.y*=nmber
        self.z*=nmber
    def calculateLength(self):
        length=((self.x**2)+(self.y**2)+(self.z**2))**(1/  2)
        return length


TwoDV1 = TwoDVector(5, 6)
TwoDV2 = TwoDVector(3, 7)
TwoDV3 = TwoDVector(1, 8)
print("===============")
TwoDV1.add2DVectors(TwoDV2, TwoDV3)
TwoDV1.print2DVector()
print("===============")
ThreeDV1 = ThreeDVector(5, 6, 1)
ThreeDV2 = ThreeDVector(1, 9, -7)
ThreeDV3 = ThreeDVector(8, 2, 4)
print("===============")
ThreeDV1.add3DVectors(ThreeDV2,ThreeDV3)
ThreeDV1.print3DVector()
print("===============")
ThreeDV1.multiplyScalar(3)
ThreeDV1.print3DVector()
print("===============")
print(ThreeDV1.calculateLength())

```

```python
class ChefsCounter:
    reservation={}
    @classmethod
    def createNewBranch(cls,branch):
        return cls(branch)
    def __init__(self,branch,res=5):
        self.branch=branch
        self.reser=res
        self.count=0
        self.name=""
        self.lst=[]
        print(f"The {branch} branch of Chef's counter is open for reservation!")
    def reserve(self,*args):
        for i in args:
            if self.count<self.reser:
                self.count+=1
                self.name+=i+", "
                self.lst.append(i)
            ChefsCounter.reservation[self.branch]=self.lst
        else:
            print(f"Sorry {args[-1]} {self.reser} people already made a reservation in this branch.")
    def reservation_info(self):
        print(f"Customers who reserved in {self.branch} branch:")
        print(self.name[:(len(self.name)-2)])

print("===============1=============")
branch1 = ChefsCounter("Gulshan")
print("===============2=============")
branch1.reserve("Sam", "Paul")
print("===============3=============")
branch1.reservation_info()
print("===============4=============")
branch1.reserve("John", "Robin", "Billy", "Robert")
print("===============5=============")
branch1.reservation_info()
print("===============6=============")
branch2 = ChefsCounter("Dhanmondi",7)
print("===============7=============")
branch2.reserve("Ben", "Alice", "Fred")
print("===============8=============")
branch2.reservation_info()
print("===============9=============")
branch2.reserve("Tom", "Ken", "Garet", "Miles", "Taylor")
print("===============10=============")
branch2.reservation_info()
print("===============11=============")
branch3 = ChefsCounter.createNewBranch("100 feet")
print("===============12=============")
branch3.reserve("Harry", "Bob", "Jenny")
print("===============13=============")
branch3.reservation_info()
print("===============14=============")
print("Reservation Information of All Branches:",ChefsCounter.reservation)
```

```python
class Ticket:
    route_distance = {"Route A":400, "Route B":425, "Route C":350}
    fare_per_km = 20
    def __init__(self, route, journeyDate, price = 0):
        self.route = route
        self.journeyDate = journeyDate
        self.__price = price
    def setPrice(self, price):
        self.__price = price
    def getPrice(self):
        return self.__price
    def ticket_details(self):
        return f"Route: {self.route}\nJourney Date: {self.journeyDate}"

class BusTicket(Ticket):
    total_tickets=0
    def __init__(self,route,date,bus,seat):
        super().__init__(route,date)
        BusTicket.total_tickets+=1
        self.bus_name=bus
        self.seat=seat
        self.status="Not Paid"
    def ticket_details(self):
        print(f"{self.bus_name}-{BusTicket.total_tickets}")
        print(super().ticket_details())
        print(self.bus_name)
        print(self.seat)
        print(self.getPrice())
        print(self.status)
    def make_payment(self):
        self.status="Paid"
        print("Payment Successfull")
    def calculate_fare(self):
        fare=Ticket.route_distance[self.route]*Ticket.fare_per_km
        self.setPrice(fare)
        print("Ticket fare is calculated successfully")

ticket1 = BusTicket("Route A", "30 April, 2023", "Nabil Enterprise", "F2")
print("Total ticket(s):", BusTicket.total_tickets)
print("1============================")
ticket1.calculate_fare()
print("2============================")
ticket1.ticket_details()
print("3============================")
ticket1.make_payment()
print("4============================")
ticket1.ticket_details()
print("5============================")
ticket2 = BusTicket("Route C", "26 April, 2023", "Hanif Enterprise", "A2")
print("Total ticket(s):", BusTicket.total_tickets)
print("6============================")
ticket2.calculate_fare()
print("7============================")
ticket2.make_payment()
print("8============================")
ticket2.ticket_details()
```

```python
class Workspace:
    def __init__(self, workspaceName, type):
        self.workspaceName = workspaceName
        self.type = type
        self.channels = {}
    def checkChannelExistence(self, channelName):
        if channelName in self.channels:
            return True
        return False
    def __str__(self):
        return f"Name: {self.workspaceName}\nType: {self.type}\n"

class SlackWorkspace(Workspace):
    total_workspaces=0
    def __init__(self,name,typ,sections=3):
        super().__init__(name,typ)
        self.sections=sections
        self.channel=0
        self.member=0
        self.lst=[]
        print(name,"is created")
        SlackWorkspace.total_workspaces+=1
     def __str__(self):
        return super().__str__()+str(self.member)+"\n"+str(len(list(self.channels.keys())))+"\n"+str(self.channels)
    def createChannels(self):
        for i in range(1,self.sections+1):
            s="Section-"+str(i)
            print(s,"channel created.")
            if s not in self.channels:
                self.channels[s]=[]
    def addMembersToChannel(self,*args):
        for i in args:
            name,section=i.split("_")
            s="Section-"+section
            if self.checkChannelExistence(s):
                print(f"{name} added to {s} channel")
                self.member+=1
                self.channels[s].append(name)
            else:
                print(f"Sorry {name}! {s} channel not found in {self.workspaceName} ")



print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
print("1=============================")
CSE111Workspace = SlackWorkspace("CSE111Workspace","Education",5)
print("2=============================")
print(CSE111Workspace)
print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
print("3=============================")
CSE111Workspace.createChannels()
print("4=============================")
print(CSE111Workspace)
print("5=============================")
john, harry, adil = "John_2", "Harry_10", "Adil_1"
alice, bob = "Alice_2","Bob_3"
print("6=============================")
CSE111Workspace.addMembersToChannel(john, harry)
CSE111Workspace.addMembersToChannel(adil)
print("7=============================")
del(CSE111Workspace.channels["Section-3"])
print(CSE111Workspace)
print("8=============================")
CSE111Workspace.addMembersToChannel(alice, bob)
print("9=============================")
print(CSE111Workspace)
print("10=============================")
CSE320Workspace = SlackWorkspace("CSE320Workspace","Education")
CSE320Workspace.createChannels()
print("11=============================")
print(CSE320Workspace)
print("12=============================")
print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
```

```python
class MetroRail:
    def __init__(self,name,stations):
        self.name=name
        self.stations=stations
        print(f"{self.name} has total {self.stations} stations.")
        self.current_station=1
        self.passengers_dict={}
    def reach_next_station(self):
        self.current_station+=1
        print(f"The train has arrived at station {self.current_station}.")
        if self.current_station in self.passengers_dict:
            print("Following passengers to get down:")
            s=""
            for i in self.passengers_dict[self.current_station]:
                s+=i+", "
            print(s[:(len(s)-2)])
            del(self.passengers_dict[self.current_station])
        else:
            print("No passengers to get down.")
        print("Current passenger list:")
        for k,v in self.passengers_dict.items():
            print(f"Station {k}: {v}")
    def board_passenger(self,*args):
        for i in range(len(args)):
            if i%2!=0:
                if args[i]<=self.stations:
                    if args[i] not in self.passengers_dict:
                        self.passengers_dict[args[i]]=[args[i-1]]
                        print(f"{args[i-1]}, welcome to {self.name}.")
                    else:
                        self.passengers_dict[args[i]].append(args[i-1])
                        print(f"{args[i-1]}, welcome to {self.name}.")
                else:
                    print(f"Sorry {args[i-1]}, the metro does not go to {args[i]}")
metro1=MetroRail("MRT Line 1",4)
print("1====================")
metro1.board_passenger("Alice",3,"Bob",3)
print("2====================")
metro1.reach_next_station()
print("3====================")
metro1.board_passenger("Harry",4)
print("4====================")
metro1.board_passenger("John",4)
print("5====================")
metro1.reach_next_station()
print("6====================")
metro1.board_passenger("Charlie",6)
print("7====================")
metro1.reach_next_station()
```
